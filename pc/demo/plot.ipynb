```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# Load data
data = np.loadtxt("data/805_full.txt", dtype=np.int8, delimiter=' ')
samples = np.loadtxt("805_8020_4006_SNP_HCLT_AG_16_100epochs_ps0.005.txt", dtype=np.int8, delimiter=' ')

print("Real data shape:", data.shape)
print("Samples shape:", samples.shape)
```

```python
# Combine
combined = np.vstack([data, samples])

# PCA
pca = PCA(n_components=2)
pcs = pca.fit_transform(combined)

# Split back
pcs_real = pcs[:data.shape[0]]
pcs_samples = pcs[data.shape[0]:]

# Plot
plt.figure(figsize=(8,6))
plt.scatter(pcs_real[:,0], pcs_real[:,1], alpha=0.6, label="Real", s=10, color="blue")
plt.scatter(pcs_samples[:,0], pcs_samples[:,1], alpha=0.6, label="Samples", s=10, color="red")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("")
plt.legend()
plt.tight_layout()
plt.show()
```
